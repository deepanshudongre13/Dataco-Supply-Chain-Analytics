### Data Cleaning - Copying & Dropping Columns

```python
df_clean = df.copy()
```

```python
drop_column = ['Customer Email','Customer Password','Customer Street','Customer Zipcode',
               'Latitude', 'Longitude',"Order Zipcode" ,'Product Description', 'Product Image', 'Product Status']

df_clean.drop(columns=drop_column, inplace=True)
```

```python
df_clean.shape
```

### COLUMN RENAMING & DATA TYPES

```python
df_clean.columns
```

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
    .str.replace("(","" ,regex = False)
    .str.replace(")","", regex = False)
                   )
```

```python
df_clean.columns
```

```python
df_clean.info()
```

```python
df_clean["order_date_dateorders"] = pd.to_datetime(df_clean["order_date_dateorders"], errors="coerce")

df_clean["shipping_date_dateorders"] = pd.to_datetime(df_clean["shipping_date_dateorders"], errors="coerce")
```

```python
id_columns = ['category_id', 'customer_id', 'department_id', 'order_customer_id',
              'order_id', 'order_item_cardprod_id', 'order_item_id', 'product_card_id', 'product_category_id']

df_clean[id_columns] = df_clean[id_columns].astype(str)
```

```python
df_clean.info()
```

## FEATURE ENGINEERING :-

### Delivery Delay

```python
df_clean[["order_date_dateorders","shipping_date_dateorders"]].head()
```

```python
df_clean["delivery_delay_days"] = (
    df_clean["shipping_date_dateorders"] - df_clean["order_date_dateorders"]
).dt.days
```

```python
df_clean["delivery_delay_days"].describe()
```

### Late Delivery Flag

```python
df_clean["is_late_delivery"] = (
    df_clean["delivery_delay_days"] > 0
).astype(int)
```

```python
df_clean["is_late_delivery"].value_counts()
```

```python
df_clean[["delivery_delay_days", "is_late_delivery"]].head(10)
```

### Shipping Efficiency

```python
df_clean["shipping_delay_vs_schedule"] = (
    df_clean["days_for_shipping_real"]
    - df_clean["days_for_shipment_scheduled"]
)
```

```python
df_clean["shipping_delay_vs_schedule"].describe()
```

### TIME-BASED FEATURES

```python
df_clean["order_year"] = df_clean["order_date_dateorders"].dt.year
```

```python
df_clean["order_month"] = df_clean["order_date_dateorders"].dt.month
```

```python
df_clean["order_days_of_week"] = df_clean["order_date_dateorders"].dt.day_name()
```

```python
df_clean[["order_year","order_month","order_days_of_week"]].head()
```

### SALES & DISCOUNT FEATURES

```python
df_clean["revenue_per_item"] = (
    df_clean["sales"] / df_clean["order_item_quantity"]
)
```

```python
df_clean["discount_flag"] = (
    df_clean["order_item_discount"] > 0
).astype(int)
```

```python
df_clean["high_quantity_flag"] = (
    df_clean["order_item_quantity"] >= 5
).astype(int)
```

```python
df_clean[[
    "sales",
    "order_item_quantity",
    "revenue_per_item",
    "order_item_discount",
    "discount_flag",
    "high_quantity_flag"
]].head()
```

### EXPORTING CLEANED DATA
